## 기업 재무 정보 crawling

### 28개 재무정보 table명 리스트 생성

```python
# fa : 재무분석
# fp : 재무상태표
# ci : 포괄손익계산서
# cf : 현금흐름표

index = -1
tbl_list = []
for i, i_kor in list(map(list, zip(['fa','fp','ci','cf'], ['재무분석','재무상태표','포괄손익계산서','현금흐름표']))):
    for j, j_kor in list(map(list, zip(['ifrs','gaap'], ['IFRS','GAAP']))):
        for k, k_kor in list(map(list, zip(['ind','con'], ['개별','연결']))):
            if (i != 'fa'):
                for l, l_kor in list(map(list, zip(['y','q'], ['연간','분기']))):
                    index += 1
                    tbl_nm = i+'_'+j+'_'+k+'_'+l
                    tbl_nm_kor = i_kor+'_'+j_kor+'_'+k_kor+'_'+l_kor
                    tbl_list.append([index, tbl_nm, tbl_nm_kor])
            else:
                index += 1
                tbl_nm = i+'_'+j+'_'+k
                tbl_nm_kor = i_kor+'_'+j_kor+'_'+k_kor
                tbl_list.append([index, tbl_nm, tbl_nm_kor])

tbl_list
```

### Crawling

```python

start_time = time.time()
# loop stock list    
for index, row in code_df.iterrows():
    
    if index <= index:
        try:
            url = 'http://media.kisline.com/fininfo/mainFininfo.nice?paper_stock='+row['code']+'&nav=4'
            source = pd.read_html(url)

            # source 내 28개 종류의 table 반복 생성
            for tbl_num, tbl_nm, tbl_nm_kor in tbl_list:

                # 임시 테이블명 df
                df = source[tbl_num]
                # crawling 결과 불필요한 컬럼 제거
                df = df[df.columns.drop(list(df.filter(regex='Unnamed')))]
                # transpose
                df = df.transpose()
                # index 컬럼 date로 변경
                df = df.rename(columns=df.iloc[0]).drop(df.index[0])
                df = df.reset_index().rename(columns={'index':'date'})
                # 주식 종목명, 코드 추가
                df.insert(loc=0, column='name', value=[row['name']]*len(df))
                df.insert(loc=1, column='code', value=[row['code']]*len(df))
                # date column format 변경
                df.date = pd.to_datetime(df.date)

                # 제무상태표의 column명이 주식종목마다 바뀌는 문제 & 상세항목 column명이 중복되는 문제 처리
                if tbl_num in range(4,12):
                    re_columns = []
                    dup_columns = []
                    for column in range(len(df.columns)):
                        # 중복컬럼인지 check
                        if df.columns[column] in df.columns.get_duplicates():
                            # 중복컬럼 중 첫번째 컬럼인지 check
                            if df.columns[column] not in dup_columns:

                                if df.columns[column] in ['매출채권및기타채권', '기타금융자산', '기타비금융자산', 
                                                          '매입채무및기타채무', '기타금융부채','기타비금융부채']:
                                    re_columns.append("비유동"+df.columns[column])
                                elif df.columns[column] in ['(대손충당금합계(-))']:
                                    re_columns.append("카드자산"+df.columns[column])
                                elif df.columns[column] in ['지배기업소유주지분']:
                                    re_columns.append("A."+df.columns[column])
                                else:
                                    re_columns.append("A."+df.columns[column])
                            else:

                                if df.columns[column] in ['매출채권및기타채권', '기타금융자산', '기타비금융자산', 
                                                          '매입채무및기타채무', '기타금융부채','기타비금융부채']:
                                    re_columns.append("유동"+df.columns[column])
                                elif df.columns[column] in ['(대손충당금합계(-))']:
                                    re_columns.append("할부금융자산"+df.columns[column])
                                elif df.columns[column] in ['지배기업소유주지분']:
                                    re_columns.append("B."+df.columns[column])
                                else:
                                    re_columns.append("B."+df.columns[column])

                            dup_columns.append(df.columns[column])

                        else:
                            re_columns.append(df.columns[column])
                    # column rename        
                    df.columns = re_columns

                # 28개 table 중 error가 발생하지 않는 곳 체크
                clear_output(wait=True)    
                print("No:{}, 종목명:{}, 종목코드:{}, table명:{}({})".
                      format(index, row['name'], row['code'], tbl_nm, tbl_nm_kor))


                # table append
                if index == 0:
                    exec("{} = df".format(tbl_nm))
                elif index > 0:
                    exec("%s = pd.concat([%s,df], axis=0, ignore_index=False, sort=False)" %(tbl_nm, tbl_nm))

            # runtime check
            clear_output(wait=True) 
            duration = time.time() - start_time
            print("No:{}, 종목명:{}, 종목코드:{}, runtime: {:d}h {:d}m {:.2f}s".
                  format(index, row['name'], row['code'], int(duration/3600), int(int(duration/60)%60), float(duration%60)))

        except ValueError as ve:
            print("ValueError: {}".format(ve))
```

## save csv file

```python
for tbl_num, tbl_nm, tbl_nm_kor in tbl_list:
    exec("{}.to_csv('{}.csv', encoding='euc-kr', index=False)".format(tbl_nm, tbl_nm_kor))
```
